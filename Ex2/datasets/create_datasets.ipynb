```python
import numpy as np
import random
import pandas as pd
import seaborn as sns
```

# line in 2d

```python
#line in 2d
xmin, xmax  = -10.0, 10.0
n = int(xmax-xmin+1)
m, q,       = 1.0, 0
noise       = .50

np.random.seed(1)
df = pd.DataFrame()
df['X1'] = np.array(np.linspace(xmin, xmax, n))
df['X2']  = [m*t  + q  for t in df.X1] 
#add some noise
df['X1']  = [t + np.random.normal(0, noise) for t in df.X1] 

df['X2']  = [t + np.random.normal(0, noise) for t in df.X2] 

df.to_csv('line2D.csv', index = False)

sns.jointplot(x= "X1", y="X2", data=df, kind="reg", color = 'g')
sns.pairplot(data=df)
```

# line in 3D


```python
#line in 2d
xmin, xmax  = -100.0, 100.0
n = int(xmax-xmin+1)
m, q,       = 1.0, 1.0
noise       = np.ogrid.0

np.random.seed(1)
df = pd.DataFrame()
df['X1'] = np.array(np.linspace(xmin, xmax, n))
df['X2'] = np.array(np.linspace(xmin, xmax, n))
df['X3']  = m*df.X1 + m*df.X2
for col in df.columns:
    df[col]  = [t + np.random.normal(0, noise) for t in df[col]] 

df.to_csv('line3D.csv', index = False)

sns.jointplot(x= "X1", y="X2", data=df, kind="reg", color = 'g')
sns.pairplot(data=df)
```

## Swiss roll


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
X, color = datasets.make_swiss_roll(n_samples=5000, noise=2.0)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
plt.show
```
