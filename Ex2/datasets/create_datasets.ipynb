```python
import numpy as np
import random
import pandas as pd
import seaborn as sns
D = 10 #dimensioni
noise = 10.0
mm = [random.randrange(-30, 30)/10  for _ in range(D)]
qq = [random.randrange(-10, 10)/10  for _ in range(D)]
X = np.linspace(-100, 100, 1000)
df = pd.DataFrame(X, columns=['X'])
for i in range(D):
    feature = 'Y'+str(i+1)
    yy = [mm[i]*xi+ qq[i]+ np.random.normal(0, noise) for xi in X]
    df[feature] = yy
#df['X'] = [xi +  np.random.normal(0, noise) for xi in X]
df['label']= [int((xi+120)/50) for xi in X]
df = df.sample(frac=1)

df.to_csv('hyperplane.csv', index = False)
sns.jointplot(x="X", y="Y2", data=df,  hue = df.label)
df['label'].value_counts()
```

## Swiss roll


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
X, color = datasets.make_swiss_roll(n_samples=5000, noise=2.0)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
plt.show
```

```python
import pandas as pd 
X, color = datasets.make_swiss_roll(n_samples=5000, noise=2.0)
df = pd.DataFrame(X, columns = ['x','y','z'])
df['label']= color
df.to_csv('roll_noise.csv', index=False)
```
